```python
import sklearn
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
print(train)
print(test)
```

```python
train = train.dropna()
test = test.dropna()
# print(train)
# print(test)
X_train = train.iloc[:,:-1].values.reshape(-1, 1)
Y_train = train.iloc[:,1:2].values.reshape(-1, 1)
X_test = test.iloc[:,:-1].values.reshape(-1, 1)
Y_test = test.iloc[:,:1].values.reshape(-1, 1)
# print(X_train)
# print(X_test)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 9))
plt.scatter(X_train, Y_train, color='blue')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
LR_model = LinearRegression()
LR_model.fit(X_train, Y_train)
print(LR_model.intercept_[0])
print(LR_model.coef_[0][0])
plt.figure(figsize=(16, 9))
plt.scatter(X_train, Y_train, color='blue')
x = np.arange(0, 100, 0.01)
y = LR_model.coef_[0][0] * x + LR_model.intercept_[0]
plt.plot(x, y, color='red')
plt.show()
```

```python
from sklearn import metrics
Y_pred = LR_model.predict(X_test)
print("MSE ", metrics.mean_squared_error(Y_test, Y_pred))
print("RMSE ", np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```
